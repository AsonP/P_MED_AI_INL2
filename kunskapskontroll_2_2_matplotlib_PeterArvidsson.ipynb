### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

Answer:
Den rekommenderade metoden är Object-Oriented Interface eftersom den ger mer flexibilitet och kontroll över plotten, särskilt när du arbetar med mer komplexa visualiseringar.

1. Procedural (Pyplot) Interface: Detta är den enklare och mer nybörjarvänliga metoden.

2. Object-Oriented Interface: Detta är den mer flexibla och kraftfulla metoden. 
Du skapar figurer och axlar som objekt och använder deras metoder för att skapa och manipulera plotten. 
Detta tillvägagångssätt ger dig mer kontroll och är bättre för komplexa plotter.


### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

Figure: Detta är hela fönstret eller bilden som innehåller alla plotter. Det är som en duk där du kan rita dina grafer.

Axes: Detta är en enskild plot eller graf inom figuren. En figur kan innehålla flera axlar.

Axis: Detta är en av de två (eller fler) skalorna som används för att rita data på axlarna. 
Det finns vanligtvis en x-axel och en y-axel.

Artist: Detta är alla objekt som kan ritas på figuren, inklusive axlar, linjer, text, etc. 
Allt som du ser på en plot är en artist.

### When plotting in Matplotlib, what is the expected input data type?

Svar:
De förväntade indata typerna vara:

NumPy Arrays: Vanligtvis används för numeriska data och ger bättre prestanda.

Pandas DataFrames: Används ofta för att hantera och analysera data i tabellform.

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
import matplotlib.pyplot as plt
import numpy as np

# Skapa data
x = np.linspace(-4, 4, 100)
y = x**2

# Skapa plotten
plt.plot(x, y)

# Lägg till titel och axel-etiketter
plt.title('Plot of y = x^2')
plt.xlabel('x')
plt.ylabel('y')

# Visa plotten
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data för scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data för bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())

# Skapa figur och subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Scatter plot
ax1.scatter(random_data_x, random_data_y, alpha=0.5)
ax1.plot(x, y, color='red')
ax1.set_title('Scatter Plot')
ax1.set_xlabel('Random Data X')
ax1.set_ylabel('Random Data Y')

# Bar plot
ax2.bar(names, values)
ax2.set_title('Bar Plot')
ax2.set_xlabel('Fruit')
ax2.set_ylabel('Quantity')

# Justera layout och visa plotten
plt.tight_layout()
plt.show()
```
